# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
total_rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fare_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. (average fare per ride is fare by type divided by total rides by type)
average_fare_per_ride = total_fare_by_city_type / total_rides_by_city_type
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. (fare by type divided by drivers by type)
average_fare_per_driver = total_fare_by_city_type / total_drivers_by_city_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_ride_summary_df = pd.DataFrame({
    "Total Rides": total_rides_by_city_type,
    "Total Drivers": total_drivers_by_city_type,
    "Total Fares" : total_fare_by_city_type,
    "Average Fare per Ride" : average_fare_per_ride,
    "Average Fare per Driver" : average_fare_per_driver})

pyber_ride_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_ride_summary_df.index.name = ""
pyber_ride_summary_df
```

```python
#  8. Format the columns. (so format with 2 decimal places since it's fares.. and then perhaps with $ signs)
pyber_ride_summary_df["Total Rides"] = pyber_ride_summary_df["Total Rides"].map("{:,}".format)
pyber_ride_summary_df["Total Drivers"] = pyber_ride_summary_df["Total Drivers"].map("{:,}".format)
pyber_ride_summary_df["Total Fares"] = pyber_ride_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_ride_summary_df["Average Fare per Ride"] = pyber_ride_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_ride_summary_df["Average Fare per Driver"] = pyber_ride_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_ride_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df.head()
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fares_by_city_date_df = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
sum_fares_by_city_date_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

sum_fares_by_city_date_df = sum_fares_by_city_date_df.reset_index()
sum_fares_by_city_date_df
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

sum_fares_by_city_date_pivot = sum_fares_by_city_date_df.pivot(index="date", columns="type", values="fare")
sum_fares_by_city_date_pivot.head(10)
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
January_to_April_Fares = sum_fares_by_city_date_pivot.loc["2019-01-01":"2019-04-28"]
January_to_April_Fares.head(10)
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

January_to_April_Fares.index = pd.to_datetime(January_to_April_Fares.index)
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
January_to_April_Fares.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
Weekly_fares_sum_df = January_to_April_Fares.resample("w").sum()
Weekly_fares_sum_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

Weekly_fares_sum_df.plot(figsize = (20,5))
# had to change fig size big enough that the legend came to the middle.. otherwise was going to have to code legend location
plt.title("Total Fare by City Type")
plt.ylabel("Fare ($USD)")
plt.xlabel("")
#above code for xlabel to get rid of label on x-axis to match the Challenge

plt.savefig("analysis/PyBer_fare_summary.png")
```
